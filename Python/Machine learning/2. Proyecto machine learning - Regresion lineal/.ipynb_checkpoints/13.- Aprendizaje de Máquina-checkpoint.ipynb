## Series de tiempo y regresión lineal simple y multiple

## Regresión lineal simple

Importamos las librerias necesarias y cargamos el dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
kchouse = pd.read_csv('Data/kc_house_data.csv')
kchouse.head()
```

Se ajustan los nombres de las columnas del dataframe

```python
kchouse.columns=['AreaFT', 'PreciosUSD']
```

Usamos train test split para training y pruebas

```python
X_train, X_test, y_train, y_test = train_test_split(kchouse.AreaFT.values.reshape(-1,1), kchouse.PreciosUSD.values, random_state=11)
```

Verificamos los tamaños de train y test

```python
X_train.shape
```

```python
X_test.shape
```

Se importa LinearRegression de sklearn

```python
from sklearn.linear_model import LinearRegression
```

Creamos el estimador y se realiza el fit

```python
reglin = LinearRegression()
```

```python
reglin.fit(X=X_train, y=y_train)
```

Obtenemos el Coef y el INtercept

```python
reglin.coef_
```

```python
reglin.intercept_
```

Para evaluar el modelo, se crean los objetos prediccion y esperado

```python
prediccion = reglin.predict(X_test)
```

```python
esperado = y_test
```

Se usa un ciclo for que muestre predicción vs esperado para los datos de 200 en 200

```python
for pred, esp in zip(prediccion[::200], esperado[::200]):
    print(f'Predicción: {pred:.2f}, Esperado: {esp:.2f}, Variación: {(pred-esp)/esp:.2%}')
```

Ahora predeciremos precios de casas usando area en sqft

```python
predice = (lambda x: reglin.coef_ * x + reglin.intercept_)
```

```python
predice(770)
```

```python
predice(1160)
```

Y ahora visualizaremos la regresion lineal con seaborn

```python
axes = sns.scatterplot(data=kchouse, x='AreaFT', y='PreciosUSD', hue='PreciosUSD')

x = np.array([min(kchouse.AreaFT.values), max(kchouse.AreaFT.values)])
y = predice(x)

line = plt.plot(x,y)
```

## Regresion lineal multiple con los datos de California

### 1. Cargar el dataset

```python
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
```

```python
print(california.DESCR)
```

```python
california.data.shape
```

```python
california.target.shape
```

```python
california.feature_names
```

### 2 Explorar los datos

```python
pd.set_option("display.precision", 4)
```

Cargamos los datos de california en un dataframe

```python
california_df = pd.DataFrame(data=california.data, columns=california.feature_names)
```

Se agrega el target de california al mismo dataframe como una columna (Serie) adicional

```python
california_df['ValorCasa'] = pd.Series(california.target)
```

Creamos a su vez un dataframe filtrado que suprima los outlayers y compararemos la diferencia con el dataset original

```python
california_filtrado = california_df

for column in list(california_df.columns):
    Q1 = california_df[column].quantile(q=0.25)
    Q3 = california_df[column].quantile(q=0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    california_filtrado = california_filtrado[(california_filtrado[column] > lower_bound) & (california_filtrado[column] < upper_bound)]
```

```python
california_df.head()
```

Usamos describe para verificar las estadisticas generales del dataset

```python
california_df.describe()
```

```python
california_filtrado.describe()
```

### 3 Visualizar los datos

Vamos a armar un muestreo del 10% del dataset para ver mas facil las visualizaciones

```python
muestradf = california_df.sample(frac=0.1, random_state=17)
```

Cambia la fuente de seaborn y colocamos el estilo Whitegrid

```python
sns.set(font_scale = 1)
```

```python
sns.set_style('whitegrid')
```

Se crean las graficas de precio vs cada caracteristica

```python
for feature in california.feature_names:
    plt.figure(figsize=(12,5))
    sns.scatterplot(data=muestradf, x=feature, y='ValorCasa', hue='ValorCasa', palette='cool', legend=False)
```

Se puede observar como la disperción de los datos se disminuye en el dataframe filtrado

```python
muestra_filtrada = california_filtrado.sample(frac=0.1,random_state=17)

for feature in california.feature_names:
    plt.figure(figsize=(12,5))
    sns.scatterplot(data=muestra_filtrada, x=feature, y='ValorCasa', hue='ValorCasa', palette='cool', legend=False)
```

### 9  Escoger el mejor modelo

```python
from sklearn.linear_model import ElasticNet, Lasso, Ridge
```

```python
estimadores = {'RegLin':LinearRegression(), 'ElasticNet':ElasticNet(), 'Lasso':Lasso(), 'Ridge':Ridge()}
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
for promest, object in estimadores.items():
    kfold = KFold(n_splits=10, random_state=11, shuffle=True)
    puntajes = cross_val_score(estimator=object, X=california.data, y=california.target, cv=kfold, scoring='r2')
    print(f'{promest:>16}:', f'Promedio de puntajes R2 = {puntajes.mean():.3f}')
```
